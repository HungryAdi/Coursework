# Problem 2

## Problem 2.1

```python
iris = np.genfromtxt("data/iris.txt",delimiter=None) # load the data
X = iris[:,0:2]
Y = iris[:,-1]

import mltools as ml

K = [1,5,10,50]

X,Y = ml.shuffleData(X,Y); # shuffle data randomly
# (This is a good idea in case your data are ordered in some pathological way,
# as the Iris data are)

Xtr,Xva,Ytr,Yva = ml.splitData(X,Y,0.75) # split data into 75/25 train/validation

for i in range(len(K)):
    knn = ml.knn.knnClassify(Xtr, Ytr, K[i]) # create the object and train it
    # where K is an integer, e.g. 1 for nearest neighbor prediction

    ml.plotClassify2D( knn, Xtr, Ytr ); # make 2D classification plot with data (Xtr,Ytr)
    plt.show()
    
    #YvaHat = knn.predict(Xva) # get estimates of y for each data point in Xva
```

## Problem 2.2

```python
K=[1,2,5,10,50,100,200]
errTrain = [0,0,0,0,0,0,0]
errTest = [0,0,0,0,0,0,0]

for i,k in enumerate(K):
    learner = ml.knn.knnClassify(Xtr, Ytr, k)
    errTrain[i] = learner.err(Xtr,Ytr)
    errTest[i] = learner.err(Xva,Yva)

# Creating subplots with just one subplot so basically a single figure.
fig, ax = plt.subplots(1, 1, figsize=(10, 8))

# I added lw (line width) and the label. 
ax.semilogx(K, errTrain, 'r-', lw=3, label='Training')
ax.semilogx(K, errTest, 'g-', lw=3, label='Validation')

# Adding a legend to the plot that will use the labels from the 'label'.
plt.legend()

# Controlling the axis.
#ax.set_xlim(0, 200)
#ax.set_ylim(0, 1)

plt.show()
```

### I would recommend a K value of 50 as it produces the lowest validation error

## Problem 2.3

```python
iris = np.genfromtxt("data/iris.txt",delimiter=None) # load the data
X = iris[:,0:4]
Y = iris[:,-1]

import mltools as ml

X,Y = ml.shuffleData(X,Y);

Xtr,Xva,Ytr,Yva = ml.splitData(X,Y,0.75)

K=[1,2,5,10,50,100,200]
errTrain = [0,0,0,0,0,0,0]
errTest = [0,0,0,0,0,0,0]

for i,k in enumerate(K):
    learner = ml.knn.knnClassify(Xtr, Ytr, k)
    errTrain[i] = learner.err(Xtr,Ytr)
    errTest[i] = learner.err(Xva,Yva)

# Creating subplots with just one subplot so basically a single figure.
fig, ax = plt.subplots(1, 1, figsize=(10, 8))

# I added lw (line width) and the label. 
ax.semilogx(K, errTrain, 'r-', lw=3, label='Training')
ax.semilogx(K, errTest, 'g-', lw=3, label='Validation')

# Adding a legend to the plot that will use the labels from the 'label'.
plt.legend()

# Controlling the axis.
#ax.set_xlim(0, 200)
#ax.set_ylim(0, 1)

plt.show()
```

### The training and test errors are much smaller and are close to 0 approaching K = 50. I would now recommend a K value of 10 as it produces the lowest validation error

 p(x4|y=-1) = 5/6 = 0.83

## Problem 3.2

x = (0 0 0 0 0) -> (1-p(x1|y=1))*(1-p(x2|y=1))*(1-p(x3|y=1))*(1-p(x4|y=1))*(1-p(x5|y=1))*(p(y=1)) = 0.25 * 1 * 0.25 * 0.5 * 0.75 * 0.4 = 0.009375

x = (0 0 0 0 0) -> (1-p(x1|y=-1))*(1-p(x2|y=-1))*(1-p(x3|y=-1))*(1-p(x4|y=-1))*(1-p(x5|y=-1))*(p(y=-1)) = 0.5 * 0.17 * 0.33 * 0.17 * 0.67 * 0.6 = 0.00192

#### Therefore the "read" class would be predicted.

x = (1 1 0 1 0) -> (p(x1|y=1))*(p(x2|y=1))*(1-p(x3|y=1))*(p(x4|y=1))*(1-p(x5|y=1))*(p(y=1)) = 0.75 * 0 * 0.25 * 0.5 * 0.75 * 0.4 = 0

x = (1 1 0 1 0) -> (p(x1|y=-1))*(p(x2|y=-1))*(1-p(x3|y=-1))*(p(x4|y=-1))*(1-p(x5|y=-1))*(p(y=-1)) = 0.5 * 0.83 * 0.33 * 0.83 * 0.67 * 0.6 = 0.0457

#### Therefore the "not read" class would be predicted
