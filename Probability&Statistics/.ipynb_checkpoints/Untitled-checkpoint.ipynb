## Question 1
* C = Cs student has Carpel Tunnel Syndrome
* P = Positive test result
\begin{equation*}
P(C) = 0.10
p(C|P) = 0.75
P(P|C) = 0.968
\end{equation*}
* The largest possible false positive rate (P(P|C')) that will reach 0.75 accuracy rate

\begin{equation*}
P(C|P) = \frac{P(P|C)P(C)}{P(P|C)P(C)+P(P|C')P(C')}\\
0.75 = \frac{0.968(0.10)}{0.968(0.10)+P(P|C')(1-0.10)}\\
0.75 = \frac{0.968(0.10)}{0.968(0.10)+P(P|C')(0.90)}\\
0.75(0.0968+P(P|C')(0.90)) = 0.0968\\
0.0968+P(P|C')(0.90) = 0.1291\\
P(P|C')(0.90) = 0.0323\\
P(P|C') = 0.0359\\
\end{equation*}

## Question 3
### 3.a
\begin{equation*}
P(B|A) > P(B)\\
\frac{P(A|B)P(B)}{P(A)} > P(B)\\
P(A|B)P(B) > P(B)P(A)\\
P(A|B) > P(A)\\
\end{equation*}
* Therefore B also repels A

### 3.b
\begin{equation*}
P(B|A) > P(B)\\
1-P(B^c|A) > 1-P(B^c)\\
-P(B^c|A) > -P(B^c)\\
P(B^c|A) < P(B^c)\\
\end{equation*}
* Therefore A repels B^c

### 3.c
\begin{equation*}
P(B|A) > P(B)\\
P(B|A) > P(B|A)P(A) + P(A^c)P(B|A^c)\\
P(B|A) > P(B|A)(1-P(A^c)) + P(A^c)P(B|A^c)\\
P(B|A) > P(B|A) -P(B|A)P(A^c) + P(A^c)P(B|A^c)\\
0 > -P(B|A)P(A^c) + P(A^c)P(B|A^c)\\
P(B|A)P(A^c) > P(A^c)P(B|A^c)\\
P(B|A) > P(B|A^c)\\
\end{equation*}

```python
from __future__ import division 
import numpy as np

trainPatient = np.load('trainPatient.npy')
testPatient = np.load('testPatient.npy')

N = trainPatient.shape[0]
Ndisease = np.sum(trainPatient[:, 2])

Nda = 0
for i in range(0,len(trainPatient)):
    if (trainPatient[i][0] >= 55):
        if (trainPatient[i][2] == 1):
            Nda += 1
Nwithout = (len(trainPatient) - Ndisease)

Nha = 0
for i in range(0,len(trainPatient)):
    if (trainPatient[i][0] >= 55):
        if (trainPatient[i][2] == 0):
            Nha += 1
            
Pdisease = Ndisease / N
Phealthy = Nwithout / N
Poverdisease = Nda/Ndisease
Poverhealthy = Nha/Nwithout
PunderPdisease = (Ndisease - Nda)/Ndisease
PunderPhealthy = (Nwithout - Nha)/Nwithout



print('Probability of heart disease:', Pdisease)
print('Probability of No heart disease: ',Phealthy)
print('Probability of age over 55 given heart disease',Poverdisease)
print('Probability of age over 55 given NO heart disease',Poverhealthy)
print('Probability of age under 55 given heart disease',PunderPdisease)
print('Probability of age under 55 given NO heart disease',PunderPhealthy)
print('\n')
```

```python
PdiseaseGunder = 0
PhealthyGunder = 0
PdiseaseGover = 0
PhealthyGover = 0
Pover = np.sum(trainPatient[:,0])/len(trainPatient)
Punder = 1 - Pover
accuracy = 0
M = testPatient.shape[0]
Yhat = np.zeros(M)

for i in range(len(testPatient)):
    if (testPatient[i][0] > 55):
        PdiseaseGover = Poverdisease * Pdisease/Pover
        PhealthyGover = Poverhealthy * Phealthy/Pover
        if (PdiseaseGover >= PhealthyGover):
            accuracy += 1
    if (testPatient[i][0] <= 55):
        PdiseaseGunder = PunderPdisease * Pdisease/Punder
        PhealthyGunder = PunderPhealthy * Phealthy/Punder
        if (PdiseaseGunder >= PhealthyGunder):
            accuracy += 1
accuracy = abs(accuracy - np.sum(testPatient[:,2]))/M
print('\nAccuracy of bayesian classifier given feature A:',accuracy)
print('\n')
```
